```python
import numpy as np
import pylab as plt
```

```python
?np.random.randn
```

```python
N = 100
np.random.seed(302)
y = np.random.randn(N) #Gaussian

#rand = uniform dist
#randn = gaussian
```

```python
plt.hist(y)
```

```python
t = np.linspace(0, N, N) #includes upper value, unlike range function
t.shape, y.shape
```

```python
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y')
plt.title('y vs t')

#This is a stochastic process because we've generated random values.
```

```python
plt.hist(y[:50], alpha = 0.4)
plt.hist(y[50:], alpha = 0.4) #alpha affects the transparency
```
